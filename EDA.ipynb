 sns.clustermap(...)

A clustermap does two things at once:

(a) Heatmap

Colors represent passenger counts

Darker/brighter colors = higher values

Shows seasonal and yearly trends visually

(b) Hierarchical clustering

Automatically groups similar rows (months) and columns (years)

Reorders rows and columns based on similarity

Adds dendrograms (tree diagrams) on top and left

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))

```
